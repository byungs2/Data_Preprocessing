```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# 요일별로 total_bill을 표현 가능, 즉 범주형, 요일별로 정산등에 유용
# 요일별 확인이 좀 더 명확하다는 장점이 있음 
sns.catplot(x = 'day', y = 'total_bill', data = tips);
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm');
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'box');
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'violin');
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'boxen');
```

```python
# 성별로 색을 구분해서 swarm 형식의 catplot 그리기
sns.catplot(x = 'day', y = 'total_bill', data = tips, hue= 'sex'  , kind = 'box');
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, hue= 'sex', kind = 'violin');
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, hue= 'sex', kind = 'violin' , split = True);
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, hue= 'sex', kind = 'violin',
            split = True, inner = 'stick');
```

```python
# ? 타이타닉 데이터를 활용해서 생존율을 확인 가능한 차트를 그려보기
titanic = sns.load_dataset('titanic')
titanic.head(10)
```

```python
sns.catplot(y = 'fare', x = 'survived' , data = titanic, hue = 'sex',kind = 'box');
```

```python
sns.catplot( x = 'survived', y = 'fare',data = titanic, hue = 'sex',kind = 'bar');
```

```python
 sns.catplot( x = 'who', y = 'survived',data = titanic,kind = 'bar', ci = None);
```

```python
# 하나의 figure에 두개의 chart(axes 두개)그리기

fig, ax = plt.subplots(1, 2, figsize = (12,5))

titanic[['sex', 'survived']].groupby(['sex']).mean().plot.bar(ax =ax[0]);
ax[0].set_title('Survived vs Sex');

sns.countplot(x = 'sex', hue = 'survived', data = titanic, ax = ax[1])
ax[1].set_title('Survived vs Sex');
```

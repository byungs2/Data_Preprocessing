```python
# 타이타닉 데이터 활용하여 생존율 확인 가능한 차트 생성
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic_surv = titanic[titanic['survived'] == 1]
titanic_surv
```

```python
titanic_dead = titanic[titanic['survived'] == 0]
titanic_dead
```

```python
sns.catplot(x='sex',y='survived', hue='class',kind='bar',data=titanic, ci = None)
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 5))
titanic[['sex', 'survived']].groupby(['sex']).mean().plot.bar(ax=ax[0])
ax[0].set_title('Rate');
sns.countplot('sex',hue = 'survived', data = titanic, ax=ax[1])
ax[1].set_title('Survived vs Dead')
```

```python
df = pd.DataFrame(dict(time=np.arange(500), value = np.random.randn(500).cumsum()))
df
```

```python
sns.relplot(x = 'time', y = 'value', data = df);
```

```python
sns.relplot(x = 'time', y = 'value', data = df, kind = 'line');
```

```python
sns.relplot(x = 'time', y = 'value', data = df, kind = 'scatter');
```

각 사람(subject)의 활동 종류(event)에 따라 각 시점(timepoint) 별로 fmri(뇌활동 측정 기술)의 측정값 중 하나의 센서값을 정리한 시계열 데이터

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
# subject 와 event를 구분 없이 timepoint 별로 반복적으로 관측된 값을 바탕으로 그림 
# 실뢰구간 과 추정 회귀선까지 그려짐
sns.relplot(x = 'timepoint', y = 'signal', kind='line', data = fmri);
```

```python
# ci = None 추가로 신뢰 구간 제거 
sns.relplot(x = 'timepoint', y = 'signal', kind='line', data = fmri, ci = None);
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind='line', data = fmri, hue = 'region');
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind='line', data = fmri, hue = 'region' , style = 'event');
```

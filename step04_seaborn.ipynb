```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.regplot(x="total_bill",y="tip", data = tips);
```

```python
sns.set(style ='darkgrid')
sns.regplot(x="total_bill",y="tip", data = tips);
```

```python
sns.relplot(x="total_bill",y="tip", data = tips, hue = "smoker");
```

```python
sns.relplot(x="total_bill",y="tip", data = tips, hue = "size");
```

```python
sns.relplot(x="total_bill",y="tip", data = tips, hue = "day");
```

```python
sns.relplot(x="total_bill",y="tip", data = tips, hue = "size", style = "smoker");
```

```python
sns.relplot(x="total_bill",y="tip", data = tips, hue = "size", style = "smoker", palette="Blues");
```

```python
sns.relplot(x="total_bill",y="tip", data = tips, hue = "size", style = "smoker", palette="Blues", size = "size");
```
